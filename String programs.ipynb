### 3.2 for loop with length function


```python
string="madam"
revstr=""

for i in range(len(string),0,-1):

  revstr=revstr+string[i-1]
  print(revstr)

if string==revstr:
  print("palindrome")
else:
  print(string,"Not palindrome")
```

## 4.while loop

```python
string="Mukesh"
revstr=""

l=len(string)

while l>0:

  revstr = revstr + string[l-1]
  l=l-1
  print(revstr)

if string==revstr:
  print(string,"is palindrome")
else:
  print(string,"is not palindrome")

```

## 5.functions


```python
def palindrome(string):

  revstr=""

  for i in range(len(string),0,-1):

    revstr=revstr+string[i-1]
    print(revstr)

  if string==revstr:
    print(string,"palindorme")
  else:
    print("Not palindrome")
palindrome("arjun")
```

## 8.Number palindrome

```python
# n=int(input("Enter number: "))

n=123
rev=0

while(n>0):

# divide the given num

    dig=n%10
    # print("dig:",dig)

    rev = (rev*10) + dig
    # print("rev",rev)

    n=n//10

print("Reverse of the number:",rev)
```

```python
import re

input_string = "hello"

# Split the input string into a list of characters
input_list = list(input_string)

# Initialize an empty output string
output_string = ""

# Iterate over the characters in the input string
for char in input_list:
    # Check if the character is a lowercase letter using regex
    if re.match(r'[a-z]', char):
        # Toggle the case of the letter and add it to the output string
        output_string += char.swapcase()
    else:
        # If the character is not a lowercase letter, add it as is
        output_string += char

print(output_string)
```

```python
import re

input_string = "hello"
output_string = ""

for match in re.finditer(r'([a-zA-Z])', input_string):
    char = match.group(0)
    if char.islower():
        output_string += char.upper()
    else:
        output_string += char.lower()

print(output_string)
```
