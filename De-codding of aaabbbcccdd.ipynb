# Removing duplicates in string

### By using set

```python
string_dup='aaaabbbbcccccddddd'
SET=set(string_dup)
b=''.join(SET)

print(SET)
print(b)
```

### By using for loop

```python
string = 'aaaabbbbcccccddddd'
string_uniq = ''
string_uniq_sort = ''

for i in string:
  if i not in string_uniq:
    string_uniq = string_uniq + i
    string_uniq_sort = i + string_uniq_sort

print('string_dup:',string)
print('string_uniq_sort:',string_uniq)
print('string_uniq_sort_rev:',string_uniq_sort)
```

# count of duplicates in string

### By using set

```python
string="abaaabbbcddcee"
SET=set(string)
print(SET)
for i in SET:
  print('count of ',i,string.count(i))
```

### By using List

```python
string="abaaabbbcddcee"
List=[]
for i in string:
  if i not in List:
    List.append(i)
print("list:",List)
for i in List:
  print("count of", i ,string.count(i))
```

### By using Dictionary

```python
string1="abaaabbbcddcee"
string2={}
for i in string1:
  # print("i",i)
  string2[i]=string2.get(i,0)+1
print(string2)
```

# Upper to Lower & Lower to Upper

```python
string1='aABbcCDd'
string2=''
for i in string1:
  if i.islower():
    string2=string2+i.upper()
  elif i.isupper():
    string2=string2+i.lower()
print('Before modification:',string1)
print('After  modification:',string2)
```

# Separate the alphabets,digits

### Alphabets,Digits

```python
string="a4c3b1d5f7"
alphabets=[]
digits=[]
for i in string:
  if i.isalpha():
    alphabets.append(i)
  else:
    digits.append(i)
output=''.join(sorted(alphabets)+sorted(digits))
print(output)
```

### Digits,Alphabets

```python
string="a4c3b1d5f7"
alphabets=[]
digits=[]
for i in string:
  if i.isalpha():
    alphabets.append(i)
  else:
    digits.append(i)
output_Not_sort=''.join(alphabets+digits)
output_sort=''.join(sorted(digits)+sorted(alphabets))
print("string:",string)
print('output_sort:',output_sort)
print('output_Not_sort:',output_Not_sort)
```

# seperate the Uppers,Lowers & Digits

```python
string='aA1bB2Cc3dD4'
U=[]
L=[]
D=[]
A=[]
for i in string:
  if i.isupper():
    U.append(i)
  elif i.islower():
    L.append(i)
  elif i.isdigit():
    D.append(i)
  elif i.isalpha():
    A.append(i)

print('string',string)
print('Uppers:',U)
print('Lowers:',L)
print('Digits:',D)
print('Alphabets:',A)
```

```python
a="a4c3b4b5n2b1j32"
alphabets=[]
digits=[]
for i in a:
  if i.isalpha():
    alphabets.append(i)
  elif i.isdigit():
    digits.append(i)

print("digits",digits)
print('alphabets',alphabets)

output1="".join(alphabets+digits)
print("output1:",output1)

output2="".join(sorted(alphabets)+sorted(digits))
print("output2:",output2)
```

# Multiple the string with digit

## A3b2c4d1e5---->AAAbbccccdeeeee

```python
a="A3b2c4d1e5"
b=""
for i in a:
  if i.isalpha():
    string=i
  else:
    digits=int(i)
    b=b+(string*digits)
print(b)
```

```python
a="A3b2c4d1e5"
for i in range(0,len(a),2):
  print(a[i]*int(a[i+1]),end="")
```

## 3a2b1c4d5f---->aaabbcddddfffff

```python
a="3A2b4c1d5e"
b=""
for i in a:
  if i.isdigit():
    digit=int(i)
  else:
    string=i
    b=b+(string*digit)
print(b)
```

```python
a="3a2b1c4d5f"
for i in range(0,len(a),2):
  print(int(a[i])*a[i+1],end="")
```

## aaabbcddddfffff ----> 3a2b1c4d5f

```python
S = 'AAAbbccccdeeeee'

R = ''
count = 1

for i in range(1,len(S)):
  if S[i] == S[i-1]:
    # print(i,':', S[i], i-1,':',S[i-1])
    count = count + 1
  else:
    R = R + S[i-1] + str(count)
    count = 1
R = R + S[-1] + str(count)
print(R)
```

### regEx


```python
import re
```

```python
pattern = 'aZ1@'

n=0
if not re.search("[a-z]",pattern):
  n+=1
if not re.search("[A-Z]",pattern):
  n+=1
if not re.search("[0-9]",pattern):
  n+=1
if not re.search("[@#$%*]",pattern):
  n+=1


if n==0:
  print("all match")
elif n==1:
  print("Not match")
```
