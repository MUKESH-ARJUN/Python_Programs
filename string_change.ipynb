```python
string="my Name is mukesh"
#O/P4:- my emaN Is hsekum
strsplit=string.split(" ")
print("strsplit:",strsplit)
revstr=[]
for i in range(0,len(strsplit)):
  print("i:",i)
  print("strsplit:",strsplit[i])
  revstr.append(strsplit[i][::-1])
print("revstr",revstr)
```

```python
string="my Name Is mukesh"
#O/P5:-my eman Is hsekum
strsplit=string.split(" ")
# print(strsplit)
revstr=[]
for i in range(0,len(strsplit)):
  if i%2==1:
    # print(strsplit[i])
    # strsplit[i][::-1]
    j = strsplit[i][::-1]
    revstr.append(j)
  else:
    # print(strsplit[i])
    revstr.append(strsplit[i])
print("revstr:",revstr)
```

```python
#O/P6:- ym name si mukesh
string="my Name Is mukesh"

strsplit=string.split(" ")
print("strsplit:",strsplit)
print("type of strsplit:",type(strsplit))

revstr=[]

for i in range(0,len(strsplit)):
  print("strsplit[i]:",strsplit[i])
  if i%2==0:
    j=strsplit[i][::-1]
    revstr.append(j)
  else:
    revstr.append(strsplit[i])
print("revstr",revstr)
```
