```python
n=int(input('enter the factorial number:'))
factorial=1

for i in range(1,n,1):   # +ve index
  factorial=factorial*i
print("factorial of the ", n ,"is:" , factorial)
```

```python
n=int(input("enter the factorial number:"))
factorial=1

for i in range(n,1,-1):   #-ve index
  factorial=factorial*i

print("factorial of the",n,"is:" ,factorial)
```

```python
def fact(n):
  if n==0 or n==1:
    return 1
  else:
    return n*fact(n-1)
print(fact(10))
```
