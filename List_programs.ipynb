# Remove the duplicate in the List

# New Section

## Method-1 (In-built functions)

```python
List = [-1,2,2,0,-3,-2,0]

remove_dup = set(List)

print('List: ',List)
print('remove_dup: ',remove_dup)
```

## Method-2 (By using conditional statements)

```python
List = [-1,2,2,0,-3,-2,0]
Remove_dup = []

for i in List:
  if i not in Remove_dup:
    Remove_dup.append(i)

print('List:',List)
print('Remove_dup: ',Remove_dup)
```

## Method-3 (By using List comprehension)

```python
List = [-1,2,2,0,-3,-2,0]
Remov_dup = []

[Remov_dup.append(i) for i in List if i not in Remov_dup]

print('List:',List)
print('Remove_dup: ',Remov_dup)
```

# Find the Largest/Smallest num in the List


### Max value in List

#### Method-1  (In-built functions)

```python
numbers = [-1, 2, 2, 0, -3, -2, 0]
print(max(numbers))
```

#### Method-2  (By using conditional statement)

```python
L = [-1, 2, 2, 0, -3, -2, 0]

max_value = L[0]
for i in range(len(L)):
  if L[i] > max_value:
    max_value = L[i]

print(max_value)
```

### Min value in the List

#### Method-1 (In-built functions)

```python
numbers = [-1, 2, 2, 0, -3, -2, 0]

print(min(numbers))
```

#### Method-2 (Conditional statement)

```python
L = [-1, 2, 2, 0, -3, -2, 0]

min_value = L[0]
for i in range(len(L)):
  if L[i] < min_value:
    min_value = L[i]

print(min_value)
```

### Find the 2nd smallest value in List

####  method -1 (In-built functions)

```python
numbers = [-1, 2, 2, 0, -3, -2, 0]
sorted_numbers = sorted(set(numbers))
print("sorted_numbers :",sorted_numbers)
second_least = sorted_numbers[1]
print('second_least :',second_least)
```

#### Method-2 ( Conditional statement)

```python
L = [-1,2,0,-3,2,0]

FH = L[0]
SH = L[0]

for i in range(len(L)):
  if L[i] > FH:
    FH = L[i]
print('First Highest Value:',FH)

for j in range(len(L)):
  if L[j] > SH and L[j] != FH:
    SH = L[j]
print('Second Highest Value',SH)
```

#### Method - 3 (List comprehension)

```python
List = [-1, 2, 2, 0, -3, -2, 0]
min_value = min(List)
second_least = min([x for x in List if x != min_value])
print(second_least)
```

#### Method -4

```python
import heapq

List = [-1, 2, 2, 0, -3, -2, 0]
second_least = heapq.nsmallest(2,set(List))[-1]
print(second_least)
```

# List elements sorting

```python

Input: ['eat', 'tea', 'bat', 'ate', 'tab', 'cat']

output: [['eat', 'tea', 'ate'], ['bat', 'tab'] , ['cat']]
```

```python

from collections import defaultdict

D =['eat', 'tea', 'bat', 'ate', 'tab', 'cat']

S = defaultdict(list)

for i in D:
  S[str(sorted(i))].append(i)

print('original List: ', D)
print('sorted List: ',S.values())

```

```python
from collections import defaultdict

L = ['eat', 'tea', 'bat', 'ate', 'tab', 'cat']
D = defaultdict(list)

for i in L:
	D[str(sorted(i))].append(i)
print(D.values())
```
