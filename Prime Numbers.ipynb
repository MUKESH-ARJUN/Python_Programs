# PRIME NUMBER B/W the RANGE

```python
IV=int(input("Initial value:"))
FV=int(input("Final value:"))
if IV==1:
  IV=IV+1
for i in range(IV,FV):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i,end=",")
```

# Count of prime numbers in the range

```python
IV=1
FV=100
List=[]
if IV==1:
  IV=IV+1
for i in range(IV,FV):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    List.append(i)
print(List)
print('count of prime No:',len(List))
```
